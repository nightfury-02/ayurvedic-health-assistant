# 00. Setup Catalog, Schema, and Volume
This notebook sets up the foundational Unity Catalog structure and simulates uploading raw files to the volume.

```
# --- Configuration ---
CATALOG_NAME = "main"
SCHEMA_NAME = "ayurgenix"
VOLUME_NAME = "files"
```

```
# Create Catalog (Usually requires admin privileges; assuming 'main' exists)
spark.sql(f"CREATE CATALOG IF NOT EXISTS {CATALOG_NAME}")

# Create Schema
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG_NAME}.{SCHEMA_NAME}")

# Create Volume
spark.sql(f"CREATE VOLUME IF NOT EXISTS {CATALOG_NAME}.{SCHEMA_NAME}.{VOLUME_NAME}")

print(f"✅ Setup complete: {CATALOG_NAME}.{SCHEMA_NAME}.{VOLUME_NAME} is ready.")
```
