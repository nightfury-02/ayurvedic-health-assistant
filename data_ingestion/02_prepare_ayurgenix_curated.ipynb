# 02 — Raw vs curated AyurGenix (chatbot-ready structured layer)

**What this notebook does**
- **`ayurgenix_dataset_raw`** — snapshot of the Delta table built in notebook 01 (audit / reprocessing).
- **`ayurgenix_curated`** — analysis-ready copy for apps; start from raw and **extend** the SQL with trims, normalised enums, or dedupe rules once you know your column names (`DESCRIBE TABLE`).

**Why two tables?** Keep a **raw** snapshot of the AyurGenix extract and a **curated** layer you can evolve (normalised dosha/symptom/herb fields, dedupe, etc.) for your **Ayurveda chatbot** and RAG metadata without losing the original row shape.

**PDFs:** notebook **03** extracts text from all Volume PDFs into **`pdf_text_chunks`** (secondary knowledge); use **`ayurgenix_curated`** as the primary structured layer.


```python
%sql
USE CATALOG ayurveda_lakehouse;
USE SCHEMA ayurgenix;
```

## Step 1 — Copy notebook-01 Delta table into a raw layer


```python
%sql
CREATE OR REPLACE TABLE ayurveda_lakehouse.ayurgenix.ayurgenix_dataset_raw AS
SELECT * FROM ayurveda_lakehouse.ayurgenix.ayurgenix_dataset;

SELECT COUNT(*) AS row_count FROM ayurveda_lakehouse.ayurgenix.ayurgenix_dataset_raw;
```

## Step 2 — Curated table for downstream use
By default this mirrors raw. Replace the `SELECT` with your normalisation (e.g. `trim` on text columns, `lower` on labels) after profiling the Kaggle CSV.


```python
%sql
CREATE OR REPLACE TABLE ayurveda_lakehouse.ayurgenix.ayurgenix_curated AS
SELECT * FROM ayurveda_lakehouse.ayurgenix.ayurgenix_dataset_raw;

SELECT * FROM ayurveda_lakehouse.ayurgenix.ayurgenix_curated LIMIT 10;
```

## Step 3 — Quick row-count parity check


```python
%sql
SELECT
  (SELECT COUNT(*) FROM ayurveda_lakehouse.ayurgenix.ayurgenix_dataset_raw) AS raw_rows,
  (SELECT COUNT(*) FROM ayurveda_lakehouse.ayurgenix.ayurgenix_curated) AS curated_rows;
```
